```python
for year in range(2015, 2025):
    globals()[f"df{year}"]["Year"] = year
```

```python

standard_columns = df2024.columns.tolist()
dfs = []
for year in range(2015, 2025):
    df = globals()[f"df{year}"]
    
    df.columns = [col.strip() for col in df.columns]
    df.columns = standard_columns[:len(df.columns)]
    dfs.append(df)

df_all_years = pd.concat(dfs, ignore_index=True)

df_all_years.sample(20)
```

```python
rare_countries = df_all_years["Country name"].value_counts()
rare_countries = rare_countries[rare_countries < 8].index

df_rare = df_all_years[df_all_years["Country name"].isin(rare_countries)]
df_rare["Country name"].value_counts()
```

```python
df_all_years["Country name"] = df_all_years["Country name"].str.strip()        
df_all_years["Country name"] = df_all_years["Country name"].str.replace(r"[^\w\s]", "", regex=True) 
```

```python
from fuzzywuzzy import process

unique_names = df_all_years["Country name"].unique()
standardized_names = {}

reference_list = df2024["Country name"].unique().tolist()


for name in unique_names:
    match, score = process.extractOne(name, reference_list)
    if score > 90:  
        standardized_names[name] = match


df_all_years["Country name"] = df_all_years["Country name"].replace(standardized_names)
```
